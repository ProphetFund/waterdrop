# Rankings

Rank addresses by amount donated, and select top 1000 for whitelist. Ensure all are EOA addresses.

```python
import pandas as pd
import numpy as np

txs_df = pd.read_csv('out/txs_df.csv')  # this is a multi GB file
txs_df["value"] = txs_df["value"].astype(np.float128)
# cast first column to string
txs_df.head()
```

```python


# let's rank the people that have contributed to this faucet
addresses = {}
for idx, tx in txs_df.iterrows():
    a_from = tx["from"]
    if a_from not in addresses:
        addresses[a_from] = 0

    value = tx["value"] / 10**18
    addresses[a_from] += value
# turn into df and order by . The first column is a string, the second is a float
addresses_df = pd.DataFrame.from_dict({'address': list(addresses.keys()), 'value': list(addresses.values())})
addresses_df = addresses_df.sort_values(by="value", ascending=False)
addresses_df.head()
```

```python
# save to csv
addresses_df.to_csv("out/addresses_ranked_df.csv")
```

```python
import web3
import json
import os
import requests

# load ../secrets.json
with open("../secrets.json") as f:
    secrets = json.load(f)

alchemy_https = secrets["alchemy_https"]
etherscan_key = secrets["etherscan_key"]

# init web3
web3 = web3.Web3(web3.HTTPProvider(alchemy_https))


def get_first_10k_transactions(
    address, net='mainnet', startblock=0, endblock=999999999, print_url=False
):
    """
    Get all transactions for an address

    """
    params = {
        "module": "account",
        "action": "txlist",
        "address": address,
        "startblock": startblock,
        "endblock": endblock,
        "sort": "desc",
        "apikey": etherscan_key,
    }
    headers = {"User-Agent": "Chrome/41.0.2228.0"}

    subdomain = "api"
    url = "https://api.etherscan.io/api"
    if net != "mainnet":
        subdomain = subdomain + "-" + net
    url = f"https://{subdomain}.etherscan.io/api"

    if print_url:
        print(url + "?" + "&".join("{}={}".format(k, v) for k, v in params.items()))
    response = requests.get(url, params=params, headers=headers)
    if response.status_code == 200:
        data = response.json()
        if data["status"] == "0":
            pass
            # print(f"Error: {data}")
        else:
            return data["result"]
    else:
        print("Error:", response.status_code)
        # print(response.text)
        return None


def is_EOA(address):
    address = web3.toChecksumAddress(address)
    code = web3.eth.getCode(address)
    # if the address contains ANY code, it's a smart contract (not an EOA)
    if code != b'':
        print(code)
        return False

    # if the address contains NO code, it could be an EOA. But let's check that it has 
    # at least one transaction
    transactions = get_first_10k_transactions(address)
    if transactions is None:
        return False
    if len(transactions) == 0:
        return False
    return True
```

```python
import time

# make a list of the top 1000 addresses that are EOAs
addresses_EOA = []
addresses_checked = 0
for idx, row in addresses_df.iterrows():
    address = row[0]
    if is_EOA(address):
        addresses_EOA.append(address)
    if len(addresses_EOA) == 1000:
        break
    addresses_checked += 1
    if addresses_checked % 10 == 0:
        print(addresses_checked, len(addresses_EOA), idx)
    time.sleep(0.5)
    
```

```python
# map EOA addresses to their donated value
addresses_EOA_dict = {}
for idx, address in enumerate(addresses_EOA):
    addresses_EOA_dict[address] = addresses[address]
```

```python
addresses_EOA_df = pd.DataFrame.from_dict({'address': list(addresses_EOA_dict.keys()), 'value': list(addresses_EOA_dict.values())})
addresses_EOA_df = addresses_EOA_df.sort_values(by="value", ascending=False)
addresses_EOA_df.tail()
```
